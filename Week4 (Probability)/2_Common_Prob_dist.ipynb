```python
# Common Probability Distributions
# Possion Distribution definition:  it is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, given that these events occur with a known constant mean rate and independently of the time since the last event.

#property of the Poisson distribution:
# 1. The mean and variance are equal.
# 2. The Poisson distribution is defined for non-negative integers (0, 1, 2, ...).
# 3. The probability of observing k events in a fixed interval is given by the formula:
#    P(X = k) = (λ^k * e^(-λ)) / k!
# Application in Machine learning:
# The Poisson distribution is often used in machine learning for modeling count data, such as the number of events occurring in a fixed period of time, like the number of emails received in an hour or the number of clicks on a website.
# Also used in natural language processing for modeling word occurrences in documents.

# Uniform distribution definition: It is a type of probability distribution in which all outcomes are equally likely. Every event has the same probability of occurring.
# Properties of the Uniform distribution:       
# 1. The mean is the midpoint of the interval.
# 2. The variance is calculated as (b - a)^2 / 12,  
#    where 'a' is the lower bound and 'b' is the upper bound of the interval.
# 3. The distribution is defined over a continuous interval [a, b].
# Application in Machine learning:
# The Uniform distribution is often used in machine learning for initializing weights in neural networks, random sampling           
# from a range of values, and in algorithms that require random number generation within a specified range.

# Visualization of distributions and understanding their properties
# Visualization helps in understand skewness, kurtosis, and outliers 
# Skewness indicates the asymmetry of the distribution,
# while kurtosis indicates the "tailedness" of the distribution.
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson, uniform , binom

# Gaussian Distribution (Normal Distribution)
x = np.linspace(-5, 5, 100)
plt.plot(x, norm.pdf(x, 0, 1), label='Normal Distribution (mean=0, std=1)')

# Binomial Distribution
n , p = 10,0.5
k = np.arange(0, n+1)   
plt.bar(k, binom.pmf(k, n, p), label='Binomial Distribution (n=10, p=0.5)', alpha=0.5   )

# Poisson Distribution
lambda_ = 3 
x = np.arange(0, 10)
plt.bar(x, poisson.pmf(x, lambda_), label='Poisson Distribution (l=3)', alpha=0.5)

plt.title("probability distribtion")
plt.legend()

```
