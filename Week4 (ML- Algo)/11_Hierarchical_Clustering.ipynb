### Hierarchical Clustering 
- Hierarchical Clustering is an unsupervised learning algorithm that builds a hierarchy of clusters. It starts with each data point as its own cluster and then merges or splits clusters based on distance measures, forming a tree like structure class a dendrogram. the hierarchy can be used to choose a suitable number of clusters by "cutting" the tree at a specific level 


```python
from scipy.cluster.hierarchy import dendrogram , linkage # scipy is a library which is used for scientific computing and it has a module called cluster which is used for clustering algorithms. linkage is a function which is used to perform hierarchical clustering and dendrogram is a function which is used to plot the dendrogram of the hierarchical clustering  
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([[1,2],[1,4],[1,0],
               [10,2],[10,4],[10,0]])
```

```python
#perform hierarchical/agglomerative clustering
z =  linkage(x, method='ward') # linkage is a function which is used to perform hierarchical clustering and method='ward' means we are using ward's method for hierarchical clustering. ward's method is a method which minimizes the variance within the cluster and maximizes the variance between the clusters. it is a popular method for hierarchical clustering
plt.figure(figsize=(8,4))
dendrogram(z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
