```python
import numpy as np
import matplotlib.pyplot as plt 

def sigmoid (z):
    return 1/(1 + np.exp(-z))
```

```python
z = np.linspace(-10,10,100)
sigmoid_val = sigmoid(z)
```

```python
plt.plot(z, sigmoid_val)
plt.title("Sigmoid function")
plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.grid()
plt.show()
```
