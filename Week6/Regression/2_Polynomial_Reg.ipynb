#### Polynomial Regression 
- polynomial regression is an extension of linear regression that modles non linear relationship by introducing higher order terms of the input feature 
--- 
- ex 
- linear regresion = b+b1x+b2y+E
- polynomial = b+b1x+b2x^2+b3x^3+......+ bnx^n +E
- the inculsion of higher order terms allows the model to capture non linear patterns in the data

#### Steps include:
- Feature transformation 
- Model training 
- Evaluation


---
- Advantage:
    - Capture non linear relationship effectively 
- Limitations:
    - Prone to overfitting with higher degree polynomials 
    - May require regularization to avoid overfitting
- Example Use case : Predicting growth pattern in biological system where relationship are non linear 


```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)
x = np.random.rand(100, 1)* 100
y = 3* x**2 +2*x+ np.random.randn(100,1) * 5
poly_feature = PolynomialFeatures(degree= 2, include_bias = False)

x_poly = poly_feature.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y )
y_pred = model.predict(x_poly)
```

```python
plt.scatter(x,y, color="blue", label="Actual Data ")
# plt.scatter(x,y_pred, color="red", label="predicted Data ")
plt.title("Polynomial regression ")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y_pred, color="red", label="predicted Data ")
plt.title("Polynomial regression ")
plt.legend()
plt.show()
```
